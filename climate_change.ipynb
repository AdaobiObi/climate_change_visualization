```python
#load dataset
df_ghg = pd.read_csv('/content/drive/MyDrive/total_ghg_data.csv')
df_ghg.head()
```

```python
#check for missing values
df_ghg.isna().any()
```

```python
#replace missing values with nan
df_ghg = df_ghg.replace("..", np.nan)
```

```python
#drop rows with missing values
df_ghg.dropna(inplace = True)
df_ghg.head()
```

```python
#drop columns in index 1 and 2 as they provide the similar information to column 0 and 3
df_ghg = df_ghg.drop(df_ghg.columns[[1, 2]],axis = 1)
df_ghg
```

```python
#clean up the column names for the years
df_ghg.rename(columns=lambda x: x.split(' ')[0], inplace=True)
```

```python
#transform datatype of the values in the years columns from object to floating number so I can plot them
df_ghg_col = df_ghg.columns[2:]
for column in df_ghg_col:
    if df_ghg[column].dtype == 'object':
        df_ghg[column] = df_ghg[column].astype(np.float32)

df_ghg.dtypes
```

```python
#sort the dataframe
df_ghg_top = df_ghg.sort_values(by=['2018'], ascending = False)
df_ghg_top.head(20)
```

```python
#plot a treemap to show the top 10 countries with the highest total greenhouse gas emissions in 2018
df_ghg_top20 = df_ghg_top[:10]
colors=['#fae588','#f79d65','#f9dc5c','#e8ac65','#e76f51']

fig = px.treemap(df_ghg_top20, path=['Country'],values='2018', width=800, height=400)
fig.update_layout(font_size = 14,treemapcolorway = colors,
                  title_text = 'Top 10 countries with highest total greenhouse gas emissions in 2018',
                  plot_bgcolor = 'rgba(0, 0, 0, 0)',paper_bgcolor = 'rgba(0, 0, 0, 0)',
                  margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
#plot a scatter plot of Total greenhouse gases emissions per country in 2018
fig = px.scatter(df_ghg_top[:20],x="Country", y="2018", size="2018",
                 size_max=60, labels = {'Country':'Country','2018':'Total Greenhouse Gases Emissions (Mt)'})
fig.update_layout({
    'plot_bgcolor': 'rgba(0, 0, 0, 0)',
    'paper_bgcolor': 'rgba(0, 0, 0, 0)',
    'title_text': 'Total Greenhouse Gas Emissions in 2018',
    'showlegend': False,
    'title_xref': 'paper',
    'title_y':0.9,
    'title_x':0.5,
    'title_xanchor': 'center',
    'title_yanchor': 'top'
})
fig.show()
```

```python
#load dataset
df_owid = pd.read_csv('/content/drive/MyDrive/owid-co2-data.csv')
df_owid.head(5)
```

```python
#filter unique country values
df_owid['country'].unique()
```

```python
#drop rows with missing values
df_owid.dropna(subset = ['total_ghg'], inplace = True)
```

```python
#filter out data per continent and for the entire world
df_world = df_owid[df_owid.country == 'World']
df_asia = df_owid[df_owid.country == 'Asia']
df_africa = df_owid[df_owid.country == 'Africa']
df_northamerica = df_owid[df_owid.country == 'North America']
df_southamerica = df_owid[df_owid.country == 'South America']
df_europe = df_owid[df_owid.country == 'Europe']
df_antarctica = df_owid[df_owid.country == 'Antarctica']
df_oceania = df_owid[df_owid.country == 'Oceania']
```

```python
#merge dataframe for all continents
df_new = pd.concat([df_asia,df_africa])
df_new = pd.concat([df_new, df_northamerica])
df_new = pd.concat([df_new, df_southamerica])
df_new = pd.concat([df_new,df_europe])
df_new = pd.concat([df_new, df_antarctica])
df_new = pd.concat ([df_new,df_oceania])
```

```python
#rename column to continent as i've filtered out the values for the continents only
df_new = df_new.rename(columns={'country': 'continent'})
```

```python
df_new.head(6)
```

```python
#plot a scatter plot showing the total greenhouse gas emission per continent per year
fig = px.scatter(df_new,x="year", y="total_ghg", size="total_ghg", color="continent",
                 log_x=True, size_max=20)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.update_layout({
'plot_bgcolor': '#efecdb',
'paper_bgcolor': 'rgba(0, 0, 0, 0)',
})
fig.show()
```

```python
#plot a line chat showing the total greenhouse gas emission per continent per year
fig = px.line(df_new,x="year", y="total_ghg", color="continent")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.update_layout({
'plot_bgcolor': 'rgba(0, 0, 0, 0)',
'paper_bgcolor': 'rgba(0, 0, 0, 0)',
})
fig.show()
```

```python
#plot circle packing chart showing total greengouse gas emission per continent
df_new1 = df_new[df_new.year==2019].sort_values('total_ghg')
circles = circlify.circlify(
    df_new1['total_ghg'].tolist(),
    show_enclosure=False,
    target_enclosure=circlify.Circle(x=0, y=0, r=1))

fig, ax = plt.subplots(figsize=(10,10))
ax.set_title('Total Greenhouse Gas Emissions per Continent')
ax.set_facecolor("#efecdb")
# Remove axes
ax.axis('off')

#find axis boundaries
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

labels = df_new1['continent']
colors=['#fae588','#f79d65','#f9dc5c','#e76f51','#FC8E8B','#F2B469']

# print circles
for circle, label, color in zip(circles, labels,colors):
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha=0.8, linewidth=2, color = color))
    plt.annotate(label, (x,y ),va='center',ha='center',size = 14)

for circle in circles:
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha=0.01, linewidth=1, fill=True))
fig.savefig('temp.png', transparent=True)
```
